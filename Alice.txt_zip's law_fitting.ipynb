```python
fp=open('Alice.txt','r',encoding='UTF-8')
line=fp.readline()
dic={}
while line:
    s=line.split()
    for x in s:
        if x not in dic:
            dic[x]=1
        else:
            dic[x]=dic[x]+1
    line=fp.readline()
fp.close()

num=[]
for key in dic:
    num.append(dic[key])
print(len(num))
num.sort()
num.reverse()
import matplotlib.pyplot as plt
plt.loglog(num)
plt.show()

a=plt.plot(num)
b=list(range(1,6332))
from scipy import optimize
def test_func(x,amp,w):
    return amp*x**w

params,par_conv=optimize.curve_fit(f=test_func,xdata=b,ydata=num)
print(params,par_conv)
plt.plot(b,test_func(b,params[0],params[1]),'r*',label='Fitted Function')
```
